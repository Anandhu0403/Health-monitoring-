```python
import matplotlib.pyplot as plt

# Accuracy values
accuracies = [0.975609756097561, 0.9672131147540983, 0.8360655737704918]

# Algorithm names
algorithms = ['MLP', 'Random Forest', 'XGBoost']

# Plotting
plt.figure(figsize=(10, 6))

# Set position of bar on X axis
r = range(len(accuracies))

# Make the plot
plt.bar(r, accuracies, color=['b', 'g', 'r'], edgecolor='grey')

# Add xticks on the middle of the group bars
plt.xlabel('Algorithms', fontweight='bold')
plt.xticks(r, algorithms)

# Add y label
plt.ylabel('Accuracy', fontweight='bold')

# Add titles and labels
plt.title('Comparison of Algorithms Accuracy')

# Show plot
plt.show()
plt.savefig('accuracy_comparison.png')
```
