```python
maps={}
def find_category_mappings(df,variable):
    return{k:i for i,k in enumerate(df[variable].unique())}
def integer_encode(df,variable,ordinal_mapping):
    df[variable]=df[variable].map(ordinal_mapping)
for variable in ["prognosis"]:
    mappings=find_category_mappings(df,variable)
    integer_encode(df,variable,mappings)
    maps[variable]=mappings
```

```python
with open('map.txt', 'w', encoding='utf-8') as f:
    f.write(str(maps))
```

```python
y=df.iloc[:,-1:]
y.head()
```

```python
x=df.iloc[:,:-1]
x.tail()
```

```python
y=df['prognosis']
from collections import Counter#checking the dataset balance
Counter(y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
xres,yres = ros.fit_resample(x, y)
```

```python
Counter(yres)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xres,yres,train_size=0.8,random_state=40)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
import joblib

# Define and train the MLP classifier
clf_mlp = MLPClassifier(hidden_layer_sizes=(10,), random_state=42)
clf_mlp.fit(xtrain, ytrain)

# Make predictions on the test set
ypred_mlp = clf_mlp.predict(xtest)



# Save the trained MLP model
joblib.dump(clf_mlp, "mlp.pkl")
```

```python
accuracy = accuracy_score(ytest, clf_mlp.predict(xtest))
print("Accuracy :", accuracy)
```

```python

# Calculate confusion matrix
conf_matrix = confusion_matrix(ytest, ypred_mlp)
print("Confusion Matrix:")
print(conf_matrix)

# Generate classification report
class_report = classification_report(ytest, ypred_mlp)
print("\nClassification Report:")
print(class_report)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot confusion matrix
plt.figure(figsize=(15, 10))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')

# Save the image as PNG
plt.savefig('confusion_matrix.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np

# Parse the classification report and plot it
class_report_dict = classification_report(ytest, ypred_mlp, output_dict=True)
class_names = list(class_report_dict.keys())[:-3]

precision = [class_report_dict[class_name]['precision'] for class_name in class_names]
recall = [class_report_dict[class_name]['recall'] for class_name in class_names]
f1_score = [class_report_dict[class_name]['f1-score'] for class_name in class_names]

x = np.arange(len(class_names))
width = 0.25

plt.figure(figsize=(15, 10))
plt.bar(x - width, precision, width, label='Precision', color='orange')
plt.bar(x, recall, width, label='Recall', color='green')
plt.bar(x + width, f1_score, width, label='F1-Score', color='blue')

plt.xlabel('Classes')
plt.ylabel('Scores')
plt.title('Classification Report Metrics')
plt.xticks(x, class_names)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Access the loss values during training
loss_values = clf_mlp.loss_curve_

# Plot the training loss
plt.figure(figsize=(10, 6))
plt.plot(loss_values, label='Training Loss', color='blue')
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initialize empty lists to store training accuracy and epochs
train_accuracy = []
epochs = []

# Define the number of epochs
num_epochs = len(loss_values)

# Iterate over each epoch
for epoch in range(1, num_epochs + 1):
    # Calculate training accuracy for the current epoch
    accuracy = clf_mlp.score(xtrain, ytrain)
    train_accuracy.append(accuracy)
    epochs.append(epoch)

# Plot the training accuracy
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_accuracy, label='Training Accuracy', color='green')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
