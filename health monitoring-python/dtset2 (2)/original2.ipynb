```python
maps={}
def find_category_mappings(df,variable):
    return{k:i for i,k in enumerate(df[variable].unique())}
def integer_encode(df,variable,ordinal_mapping):
    df[variable]=df[variable].map(ordinal_mapping)
for variable in ["prognosis"]:
    mappings=find_category_mappings(df,variable)
    integer_encode(df,variable,mappings)
    maps[variable]=mappings
```

```python
maps
```

```python
with open('map.txt', 'w', encoding='utf-8') as f:
    f.write(str(maps))
```

```python
y=df.iloc[:,-1:]
y.head()
```

```python
x=df.iloc[:,:-1]
x.tail()
```

```python
y=df['prognosis']
from collections import Counter#checking the dataset balance
Counter(y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
xres,yres = ros.fit_resample(x, y)
```

```python
Counter(yres)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xres,yres,train_size=0.8,random_state=40)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
import joblib

# Define and train the MLP classifier
clf_mlp = MLPClassifier(hidden_layer_sizes=(10,), random_state=42)
clf_mlp.fit(xtrain, ytrain)

# Make predictions on the test set
ypred_mlp = clf_mlp.predict(xtest)



# Save the trained MLP model
joblib.dump(clf_mlp, "mlp.pkl")
```

```python
accuracy = accuracy_score(ytest, clf_mlp.predict(xtest))
print("Accuracy :", accuracy)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import joblib
from sklearn.model_selection import train_test_split

# Assuming you have your data xtrain, xtest, ytrain, ytest
# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)  # Adjust test_size as needed

# Define and train the Random Forest classifier with reduced parameters
clf_rf = RandomForestClassifier(n_estimators=10, max_features=0.5, min_samples_split=5, random_state=42)  # Adjust parameters
clf_rf.fit(xtrain, ytrain)

# Make predictions on the test set
ypred_rf = clf_rf.predict(xtest)

# Print accuracy before reducing it
print("Accuracy:", accuracy_score(ytest, ypred_rf))

# Save the trained Random Forest model
joblib.dump(clf_rf, "random_forest.pkl")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, f1_score, recall_score
import joblib

# Load the trained Random Forest model
clf_rf = joblib.load("random_forest.pkl")

# Make predictions on the test set
ypred_rf = clf_rf.predict(xtest)

# Calculate and print the accuracy score
accuracy = accuracy_score(ytest, ypred_rf)
print("Accuracy:", accuracy)

# Calculate and print the confusion matrix
cm = confusion_matrix(ytest, ypred_rf)
print("Confusion Matrix:")
print(cm)

# Calculate and print the classification report
cr = classification_report(ytest, ypred_rf)
print("Classification Report:")
print(cr)

# Calculate and print precision, recall, and F1-score
precision = precision_score(ytest, ypred_rf, average='weighted')
recall = recall_score(ytest, ypred_rf, average='weighted')
f1 = f1_score(ytest, ypred_rf, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
